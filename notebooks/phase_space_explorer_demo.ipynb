# Phase Space Explorer — Demo

Interactive exploration of dynamical systems. Choose a system, set parameters and initial conditions with sliders, then simulate and view dimension-aware plots (trajectory, vector field, equilibria).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

# Import the package (works from project root or from notebooks/ directory)
import sys
from pathlib import Path
root = Path.cwd() if (Path.cwd() / "phase_space_explorer").is_dir() else Path.cwd().parent
sys.path.insert(0, str(root))
from phase_space_explorer import (
    ALL_SYSTEMS,
    simulate,
    plot_system,
    get_equilibria_with_stability,
)
from phase_space_explorer.ui import sliders_for_system, params_from_sliders, y0_from_sliders
```

## Quick view: select system and run with default parameters

Pick a system from the dropdown and run the cell to see one trajectory with default parameters.

```python
def run_default(system_name):
    system = next(s for s in ALL_SYSTEMS if s.name == system_name)
    params = {k: info["default"] for k, info in system.parameter_info.items()}
    y0 = [info["default"] for info in system.initial_condition_info.values()]
    t_span = (0, 20) if system.dimension < 3 else (0, 30)
    t_eval = np.linspace(t_span[0], t_span[1], 500)
    t, y = simulate(system, y0, params, t_span, t_eval=t_eval)
    equilibria = get_equilibria_with_stability(system, params)
    plot_system(system, params, t, y, equilibria=equilibria)
    plt.show()

interact(run_default, system_name=Dropdown(
    options=[s.name for s in ALL_SYSTEMS],
    description="System",
));
```

## 1. Logistic growth (1D)

Sliders: parameters `r`, `K` and initial condition `x`. Plots: trajectory *x(t)*, vector field *f(x)*, phase line with equilibria (green = stable, red = unstable).

```python
from phase_space_explorer import LOGISTIC

param_s, ic_s = sliders_for_system(LOGISTIC)

def explore_logistic(**kwargs):
    params = {k: kwargs[k] for k in LOGISTIC.parameter_info}
    y0 = [kwargs[k] for k in LOGISTIC.initial_condition_info]
    t_span = (0, 20)
    t_eval = np.linspace(0, 20, 500)
    t, y = simulate(LOGISTIC, y0, params, t_span, t_eval=t_eval)
    equilibria = get_equilibria_with_stability(LOGISTIC, params)
    plot_system(LOGISTIC, params, t, y, equilibria=equilibria)
    plt.show()

interact(explore_logistic, **{**param_s, **ic_s});
```
